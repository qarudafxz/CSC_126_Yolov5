<h2>Step 2: Prepare Your Dataset</h2>

**2. Define Data Configuration:**
Edit the data.yaml file in the yolov5/data directory to define your dataset. Specify the number of classes, train and validation directories, and class names.

<h2>Step 4: Train YOLOv5</h2>

**1. Start Training:**
Run the training script with the following command:

```python
! python CSC_126_Yolov5/train.py --img 640 --batch 2 --epochs 100 --data CSC_126_Yolov5/data/data.yaml --cfg CSC_126_Yolov5/models/yolov5s.yaml --weights ''
```

* --img: Input image size.
* --batch: Batch size.
* --epochs: Number of training epochs.
* --data: Path to your data.yaml file.
* --cfg: Path to your model configuration YAML file.
* --weights: Path to pre-trained weights (if any).
